```python
import os
import pandas as pd
import plotly.express as px
from google.colab import files
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
```

```python
import plotly.graph_objects as go
```

```python
from jupyter_dash import JupyterDash
```

```python
# Upload the dataset manually
uploaded = files.upload()

# Load the dataset (ensure the file name matches what you upload)
file_path = "Updated_Impacts_of_Fireworks_with_Weights.csv"  # Adjust if file name differs
dataset_visualization = pd.read_csv(file_path)
```

```python
# Create a professional sunburst chart with advanced visuals
fig = px.sunburst(
    dataset_visualization,
    path=['Category', 'Regions Affected'],
    values='Weight',  # Use the weight column for proportional representation
    color='Weight',
    color_continuous_scale=px.colors.qualitative.Bold,  # Updated to a better professional palette
    hover_data=['Quantified Impact', 'Data Source'],
    title="Global Impacts of New Year Fireworks on Wildlife, Air, and Water"
)

# Update layout for a polished professional appearance
fig.update_layout(
    title=dict(
        text="Impacts of New Year Fireworks on Wildlife, Air, and Water",
        x=0.5,  # Center-align the title
        font=dict(size=34, color='navy', family="Arial Black")  # Bold, dark blue font
    ),
    margin=dict(t=100, l=70, r=70, b=50),  # Adjust margins for better spacing
    hoverlabel=dict(
        font_size=16,  # Increase hover text size
        font_family="Arial",
        bgcolor="#444444",  # Neutral dark gray background for hover labels
        font_color="white"  # White hover text
    ),
    paper_bgcolor="white"  # Clean white background
)

# Adjust trace properties for improved text visibility inside the chart
fig.update_traces(
    textinfo="label+percent entry+value",  # Show labels, percentages, and values
    insidetextorientation="horizontal",  # Horizontal text for readability
    textfont=dict(
        size=28,  # Increase text size for professional look
        color="white"  # Dark text for contrast
    ),
    marker=dict(line=dict(color='white', width=1))  # Add a light border for clarity
)

# Display the visualization directly in the Colab output
fig.show()

# Save the interactive visualization to /content
output_file = "Polished_Professional_Visualization_With_Weights.html"
fig.write_html(output_file)

# Download the HTML file
files.download(output_file)

print(f"Visualization saved successfully and ready for download: {output_file}")
```

```python
# Create a 3D scatter plot to represent the impacts of fireworks
fig = go.Figure()

# Add data points for the visualization
fig.add_trace(go.Scatter3d(
    x=dataset_visualization['Weight'],
    y=dataset_visualization['Category'],
    z=dataset_visualization['Regions Affected'],
    mode='markers',
    marker=dict(
        size=dataset_visualization['Weight'] / 10,  # Scale marker size by weight
        color=dataset_visualization['Weight'],  # Use weight for color gradient
        colorscale='Viridis',  # Professional color scale
        opacity=0.8
    ),
    text=dataset_visualization['Quantified Impact'],  # Add hover text
    hoverinfo='text'
))

# Update layout for professional appearance
fig.update_layout(
    scene=dict(
        xaxis_title='Weight',
        yaxis_title='Category',
        zaxis_title='Regions Affected',
        xaxis=dict(backgroundcolor="white"),
        yaxis=dict(backgroundcolor="white"),
        zaxis=dict(backgroundcolor="white"),
    ),
    title=dict(
        text="3D Impacts of New Year Fireworks on Wildlife, Air, and Water",
        x=0.5,  # Center-align the title
        font=dict(size=24, color='darkblue', family="Arial Black")
    ),
    margin=dict(t=50, l=50, r=50, b=50),  # Adjust margins
    hoverlabel=dict(
        font_size=14,
        font_family="Arial",
        bgcolor="#333333",  # Dark hover label background
        font_color="white"  # White hover text
    ),
    paper_bgcolor="white"  # White background for the plot
)

# Display the visualization
fig.show()
```

```python
# Create a 3D scatter plot to represent the impacts of fireworks
fig = go.Figure()

# Add data points for the visualization
fig.add_trace(go.Scatter3d(
    x=dataset_visualization['Weight'],
    y=dataset_visualization['Category'],
    z=dataset_visualization['Regions Affected'],
    mode='markers+text',  # Add markers and text labels
    marker=dict(
        size=dataset_visualization['Weight'] / 5,  # Dynamic scaling for better proportionality
        color=dataset_visualization['Weight'],  # Use weight for color gradient
        colorscale='Turbo',  # High-contrast, professional color scale
        opacity=0.8
    ),
    text=dataset_visualization['Quantified Impact'],  # Text labels for hover
    hoverinfo='text+x+y+z'  # Show detailed hover info
))

# Update layout for professional appearance
fig.update_layout(
    scene=dict(
        xaxis=dict(
            title='Weight',
            gridcolor="lightgray",
            zerolinecolor="gray",
            backgroundcolor="white"
        ),
        yaxis=dict(
            title='Category',
            gridcolor="lightgray",
            zerolinecolor="gray",
            backgroundcolor="white"
        ),
        zaxis=dict(
            title='Regions Affected',
            gridcolor="lightgray",
            zerolinecolor="gray",
            backgroundcolor="white"
        ),
    ),
    title=dict(
        text="3D Impacts of New Year Fireworks on Wildlife, Air, and Water",
        x=0.5,  # Center-align the title
        font=dict(size=24, color='darkblue', family="Arial Black")
    ),
    margin=dict(t=50, l=50, r=50, b=50),  # Adjust margins
    hoverlabel=dict(
        font_size=14,
        font_family="Arial",
        bgcolor="#444444",  # Neutral dark gray for hover labels
        font_color="white"  # White hover text
    ),
    paper_bgcolor="white",  # Clean white background
    showlegend=False  # Remove legend for a cleaner look
)

# Add annotations for key insights (example)
fig.add_trace(go.Scatter3d(
    x=[max(dataset_visualization['Weight'])],
    y=["Air Quality Degradation"],  # Example category
    z=["Europe"],  # Example region
    mode='text',
    text=["Highest Impact"],  # Annotation text
    textfont=dict(
        size=12,
        color='red'
    )
))

# Display the visualization
fig.show()
```
